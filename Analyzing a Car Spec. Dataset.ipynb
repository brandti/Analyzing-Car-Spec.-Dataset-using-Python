```python
# Fit the data to create a linear regression model where the target is ‘CO2 Emissions (y)’ and the feature is ‘Fuel Consumption (x)’.
import matplotlib.pyplot as plt
from scipy import stats

dataframe = pd.read_csv("C:\\Users\Mariana\Documents\BELLA SENIOR CLASSES\Sample.csv")
x = dataframe.FUEL_CONSUMPTION
y = dataframe.CO2_EMISSIONS

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):  
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x,y, color='black')
plt.plot(x, mymodel)
plt.title('Linear Regression')
plt.xlabel('Fuel Consumption')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
#importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#reading and loading the dataset
df = pd.read_csv("C:\\Users\Mariana\Documents\BELLA SENIOR CLASSES\Sample.csv")
df.head()
```

```python
#recreating the scatterplot
plt.scatter(df['FUEL_CONSUMPTION'],df['CO2_EMISSIONS'])
```

```python
#redefining the x variable as fuel consumption AND engine size so we can predict the emissions based on both
X = df[['FUEL_CONSUMPTION','ENGINE_SIZE']]
Y = df['CO2_EMISSIONS']
X.head(10)
```

```python
#splitting the data into test/train set
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2)
x_train
```

```python
#create object of the LinearRegression class
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
#fitting the x and y train variables
clf.fit(x_train,y_train)
```

```python
#predicting output using x_test variable
clf.predict(x_test)
```

```python
#test the model accuracy
clf.score(x_test,y_test)
#our model is accurate 93% 
```
